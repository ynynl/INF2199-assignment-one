```python
# import these python libraries

# data libraries
import pandas as pd
import requests

# visualization libraries
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

# set the standard plot size
plt.rcParams['figure.figsize']=(15,15)
sns.set()
```


# Get Some Data From the City of Toronto

We're going to use the "Covid-19 Cases in Toronto" dataset from the City's open data portal:
https://open.toronto.ca/dataset/covid-19-cases-in-toronto/


```python
# Get the dataset metadata by passing package_id to the package_search endpoint
# For example, to retrieve the metadata for this dataset:

url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/package_show"
params = { "id": "64b54586-6180-4485-83eb-81e8fae3b8fe"}
package = requests.get(url, params = params).json()
# print(package["result"])
```

```python
# Get the data by passing the resource_id to the datastore_search endpoint
# See https://docs.ckan.org/en/latest/maintaining/datastore.html for detailed parameters options
# For example, to retrieve the data content for the first resource in the datastore:

for idx, resource in enumerate(package["result"]["resources"]):
    if resource["datastore_active"]:
        url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search"
        p = { "id": resource["id"] }
        data = requests.get(url, params = p).json()
        df = pd.DataFrame(data["result"]["records"])
        break
        
# CKAN stands for Comprehensive Knowledge Archive Network
```

```python
df.head(20)
```


# Let's Make Some Visualizations!


```python
fig = plt.figure()
fig.suptitle('Toronto COVID-19 Source of Infection', fontsize=25) # Add the text/suptitle to figure

ax = df['Source of Infection'].value_counts().plot(
    kind='bar', 
    figsize=(14,8)
)
ax.set_xlabel("Source of Infection", fontsize=16)
ax.set_ylabel("Count", fontsize=16)
ax.tick_params(axis='both', labelsize=14)

# if you want to save your graph
fig.savefig('images/infection.jpg')
```

```python
# here are a handful of the various functions you'll use with matplotlib 
plt.figure() # create a new figure
plt.plot() # plot y versus x as lines and/or markers
plt.xlabel() # set the label for the x axis
plt.ylabel() # set the label for the y axis
plt.title() # set a title for the axes
plt.grid() # configure grid lines
plt.legend() # place a legend on the axes
plt.savefig() # save the current figure
plt.show() # display a figure
plt.clf() # clear the current figure
```

```python
# let's try seaborn, a declarative front-end to matplotlib
# this distribution plot uses a histogram-like method to create bars
# but we'll cheat a bit by using it since seaborn's default bar method wants averages and i don't want to go prep the data

g = sns.displot(df, x="Source of Infection")
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(15,15) # and we'll use this to change the figure size
```


# Something to Keep in Mind...

There is a difference between bar charts and histograms. With bar charts, a column represents a group defined by a categorical variable. With histograms, a column represents a group defined by a continuous, quantitative variable. I'm being a bit pedantic, as the methods I'm using here really only come down to whether or not I've re-shaped my original dataframe. (I could "follow the rules" by simply creating a new dataframe with a column for age categories and a column for their respective counts, but cheating using methods designed for histograms and distributions is fine for today.) 

Here's the important point: **Having a sense of what kind of visual output I'm aiming for will determine how much time I spend re-shaping my original table/spreadhsheet data.**


```python
# anyway, let's do some comparisons. 
# matplotlib doesn't handle multivariate visualization very easily, so we'll use seaborn's "dodge" option

g = sns.displot(df, x="Source of Infection", hue="Client Gender", multiple="dodge")
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(20,10)
```

```python
# side-by-side comparison is also easy
g = sns.displot(df, x="Source of Infection", hue="Client Gender", multiple="dodge", col="Client Gender")
g.set_xticklabels(rotation=90)
sns.set(font_scale=1.3) # seaborn's method for setting fontsize is pretty dumb
g.fig.set_size_inches(20,10)
```

```python
# seaborn's countplot method is a good alternative to displot
# we can use it to show the counts of observations in each categorical bin
# remember, that seaborn's barplot method, by default, sets bar height with averages
plt.figure(figsize=(20,10))
g = sns.countplot(data=df, x='Source of Infection', palette='colorblind')
```


# What Are We Missing?

Are any of you wondering why the counts are so low? By default, CKAN's default download is 100 rows. There have been around 85,000 cases of COVID-19 in Toronto, so the visualizations we've created up to now would be useless! Be careful with your data sources, especially if the API is provided with little documentation!! Of course, we could have avoided this mess by downloading the .csv and using pd.read_csv() to create our dataframe, but we'd run the risk that our data would get stale.

Anyway, let's go back up to cell 3 and add ?limit=100000 to the end of the url so that it looks like this: https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search?limit=100000


```python
# now re-import the data 
for idx, resource in enumerate(package["result"]["resources"]):
    if resource["datastore_active"]:
        url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search?limit=100000" # with updated url
        p = { "id": resource["id"] }
        data = requests.get(url, params = p).json()
        full_df = pd.DataFrame(data["result"]["records"]) # and make sure you give your full dataframe a different variable name so you don't run into state issues
        break
```

```python
full_df.head()
```

```python
g = sns.countplot(x="Age Group", hue="Source of Infection", data=full_df)
g.set_xticklabels(g.get_xticklabels(),rotation=90);
```

```python
# hey look... it's a donut chart!
# basically, make a pie chart
ax = full_df['Source of Infection'].value_counts().plot(kind='pie', figsize=(10,10))
ax.set_ylabel('')
 
# then add a circle at the center
my_circle=plt.Circle( (0,0), 0.5, color='white') # try playing around with the inner diameter
p=plt.gcf()
p.gca().add_artist(my_circle)
 
plt.show()
```

```python
# and let's add some interaction using plotly
# like seaborn, it's a simple declarative front end that makes it really easy to switch options/parameters
# also has a great built-in chart editor tool

import plotly.graph_objects as go

fig = go.Figure(go.Bar(
            y=full_df['Age Group'].value_counts()))
fig.show()

# still lots to do to get this visualization ready (order, labels, colour, etc.)
# but the interaction tools are why we will use plotly throughout the course
```

```python
# we could also use altair, but it is not as supported
# https://altair-viz.github.io/

import altair as alt

alt.Chart(df).mark_bar().encode( # notice that I switched back to df from full_df... altair has a 5000 row max
    alt.X("Age Group"),
    y='count()',
    color='Source of Infection', # could switch between gender, outcome, neighbourhood, source of infection, etc...
    tooltip=['Age Group', 'Client Gender', 'Outcome']
).interactive().properties(
    width=600, height=400
).configure_axis(
    labelFontSize=18,
    titleFontSize=18
)
```
